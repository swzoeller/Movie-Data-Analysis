```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
from config import omdb_api_key
import numpy as np
```

```python
def get_info(soup):
    '''info = {}
    title_link = soup.find('h3').find('a')
    info['id'] = title_link['href'].split('/')[2]
    info['title'] = title_link.get_text(strip=True)
    return info'''
    return soup.find('h3').find('a')['href'].split('/')[2]
```

```python
def get_company_movie_ids(company_id):
    start = 1
    next_tag = True
    results = []
    while next_tag:
        print(start)
        soup = BeautifulSoup(requests.get(f'https://www.imdb.com/search/title/?companies={company_id}&sort=release_date,asc&start={start}').content, 'html.parser')
        for item in soup.find_all('div', class_='lister-item mode-advanced'):
            results.append(get_info(item))
        next_tag = soup.find('a', class_='lister-page-next')
        start += 50
    return results
```

```python
stream_companies = [
    {'name':'Netflix Studios', 'id':'co0650527'},
    {'name':'Amazon Studios', 'id':'co0319272'},
    {'name':'Hulu Originals', 'id':'co0687042'}
]

big_companies = [
    {'name':'Warner Bros. Entertainment', 'id':'co0080422'},
    {'name':'Columbia Pictures', 'id':'co0050868'},
    {'name':'Hulu Originals', 'id':'co0687042'}
]
```

```python
results = []
for company in stream_companies:
    print(company['name'])
    company_results = get_company_movie_ids(company['id'])
    '''for company_result in company_results:
        company_result['company_id'] = company['id']
        company_result['company_name'] = company['name']'''
    results.extend(company_results)
```

```python
entries = []
for i, result in enumerate(results):
    print(i, result)
    response = requests.get(f'http://www.omdbapi.com/?apikey={omdb_api_key}&i={result}')
    entries.append(response.json())
df = pd.DataFrame(entries)
```

```python
def get_budget(title):
    try:
        return int(list(budget_df.loc[budget_df['movie']==title, 'production_budget'])[0][1:].replace(',',''))
    except IndexError:
        return np.nan
```

```python
movies_df['tn_production_budget'] = movies_df.Title.map(get_budget)
```
